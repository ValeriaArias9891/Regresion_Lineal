```python
import matplotlib.pyplot as plt  #Se importa la libreria de matplotlib.pyplot que permitirá hacer el ploteo de las gráficas
from sklearn import linear_model # Se importa la función linear_model que estima los coeficientes de esparcimiento
import seaborn as sns  # Importamos la ibreria Seaborn que nos permite hacer la visualización de datos estadísticos
import numpy as np # Importamos la libreria numpy que permite la utilización y creación de vectores y matrices 
import pandas as pd # Importamos la libreria pandas que permite el análisis y manipulación de los datos estadísticos 
```

```python
df.plot.scatter(x="Pregnancies", y="Glucose") # En éste bloque se hace la graficación  de age vs survived
plt.show()# muestra la grafica
```

```python
regresion = linear_model.LinearRegression() # se crea la línea de regresión
```

```python
pregnacies = df["Pregnancies"].values.reshape((-1,1))# crea un array de la columna pregnancies y la redimensiona con la función reshape
```

```python
modelo = regresion.fit(pregnacies, df["Glucose"])
```

```python
print("Interseccion (b)", modelo.intercept_)
print("Pendiente (m)", modelo.coef_)
```

```python
entrada = [[0],[1],[2],[5],[8],[10]]
predicciones = modelo.predict(entrada)
print(predicciones)
```

```python
df.plot.scatter(x="Pregnancies", y="Glucose", label='Datos')
plt.scatter(entrada, predicciones, color= 'pink')
plt.plot(entrada, predicciones, color= 'black', label = 'Linea de regresión')
plt.xlabel('Embarazos')
plt.ylabel('Glucosa')
plt.legend()
plt.show()
```
